1.	The boys in the class are having mean weight= 9.5kg
Standard deviation=1.1kg
Calculate the percentage(%) of the boys in the following:
1.	Weight less than 8.4kg
2.	Between 7.3kg and 11.7kg
3.	More than 12.8


```python
import scipy.stats as stats
```

```python
stats.norm.cdf(8.4,loc=9.5,scale=1.1)
```

```python
p1= stats.norm.cdf(7.3, loc=9.5,scale=1.1)
p1
```

```python
p2= stats.norm.cdf(11.7, loc=9.5,scale=1.1)
p2
```

```python
p2-p1
```

```python
p3= stats.norm.cdf(12.8, loc=9.5,scale=1.1)
```

```python
1-p3
```

```python
pop_data= stats.norm.rvs(size=100,loc=50,scale=70)
```

```python
pop_data
```

```python
pop_mean= pop_data.mean()
pop_mean
```

```python
samp_data=  stats.norm.rvs(size=70,loc=40,scale=60)
```

```python
samp_data
```

```python
stats.ttest_1samp(samp_data,popmean=pop_mean)
```

```python
data1= stats.norm.rvs(size=100,loc=40,scale=60)
```

```python
data1
```

```python
data2= stats.norm.rvs(size=100,loc=50,scale=80)
data2
```

```python
stats.ttest_ind(data1,data2,equal_var=False)
```

```python
befor_treat= stats.norm.rvs(size=100, loc=50,scale=60)
```

```python
befor_treat
```

```python
after_treat= befor_treat + stats.norm.rvs(size=100, loc=20,scale=40)
```

```python
stats.ttest_rel(befor_treat, after_treat)
```
